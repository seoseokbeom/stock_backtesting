```python
import yfinance as yf
import ta
import pandas as pd
from datetime import date, timedelta, datetime
from IPython.display import clear_output
import matplotlib.pyplot as pp
import plotly.graph_objects as go

```

```python
def get_stock_backtest_data(ticker, start_date, end_date):
    date_fmt = '%Y-%m-%d'
    start_date_buffer = datetime.strptime(
        start_date, date_fmt) - timedelta(days=365)
    start_date_buffer = start_date_buffer.strftime(date_fmt)

    df = yf.download(ticker, start=start_date_buffer, end=end_date)

    return df
```

```python
df = get_stock_backtest_data('TQQQ', '2019-01-01', '2021-01-31')
df
```

```python
QQQ = yf.Ticker('QQQ')
```

```python
old  =  QQQ.history(start="2010-01-01",  end=datetime.today())
old.head()
```

```python
old = old.reset_index()
for i in ['Open', 'High', 'Close', 'Low']: 
      old[i]  =  old[i].astype('float64')
```

```python
old
```

```python
# import plotly.graph_objects as go
fig = go.Figure(data=[go.Candlestick(x=old['Date'],
                                   open=old['Open'],
high=old['High'],
low=old['Low'],
close=old['Close'])])
fig.update_layout(xaxis_rangeslider_visible=False)
fig.show()
```

```python
Roll_Max = old['Close'].cummax()
Daily_Drawdown = old['Close']/Roll_Max - 1.0
# Daily_Drawdown
Max_Daily_Drawdown = Daily_Drawdown.cummin()
Max_Daily_Drawdown
```

```python
window = 252

# Calculate the max drawdown in the past window days for each day in the series.
# Use min_periods=1 if you want to let the first 252 days data have an expanding window
Roll_Max = old['Close'].rolling(window, min_periods=1).max()
Daily_Drawdown = old['Close']/Roll_Max - 1.0

# Next we calculate the minimum (negative) daily drawdown in that window.
# Again, use min_periods=1 if you want to allow the expanding window
Max_Daily_Drawdown = Daily_Drawdown.rolling(window, min_periods=1).min()
Max_Daily_Drawdown
# Plot the results
Daily_Drawdown.plot()
Max_Daily_Drawdown.plot()
pp.show()
```

```python
print('MDD:',round(-min(Max_Daily_Drawdown)*100,2),'%')
```
